## 3. LLM 답변 생성
- OpenAI 대시보드에서 발급받은 API Key를 OPENAI_API_KEY라고 저장하면 별도의 설정 없이 ChatOpenAI를 사용할 수 있음

```python
from langchain_openai import ChatOpenAI

llm = ChatOpenAI(model_name='gpt-4o-mini')
```

- Google AI Studio 에서 발급받은 API Key를 GOOGLE_API_KEY라고 저장하면 별도의 설정 없이 GOOGLE_API_KEY를 사용할 수 있음

```python
from langchain_google_genai import ChatGoogleGenerativeAI

llm = ChatGoogleGenerativeAI(
    model="gemini-2.5-flash"
)
```

```python
ai_message = llm.invoke("인프런에 어떤 강의가 있나요?")
```

```python
ai_message.content
```

[참고] 다른 이름으로 환경 변수를 설정할 경우
- 하지만 기본적으로 권장하는 환경 변수 이름으로 설정 추천

```python
import os
from langchain_google_genai import ChatGoogleGenerativeAI
api_key = os.getenv("API_KEY")

llm = ChatGoogleGenerativeAI(
    model="gemini-2.5-flash",
    api_key = api_key
)
```
